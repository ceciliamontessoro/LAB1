```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS

import matplotlib as mpl
import matplotlib.pyplot as plt


mpl.rcParams['font.size']=12                #10 
mpl.rcParams['savefig.dpi']=100             #72 

stopwords = set(STOPWORDS)
```

```python
# First part of the exercise
# First we load the dataset and take a look at it


wr=pd.read_csv("./dataset/winemag-data_first150k.csv")
wr.head()
wr["description"][0]
```

```python
plt.figure(figsize=(13,11))

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=60, 
                          random_state=42,
                          colormap="tab10"
                         ).generate(str(wr['description'][0]))

plt.imshow(wordcloud)
plt.title("test", fontsize=40)
plt.axis('off')
plt.show()
#fig.savefig("word1.png", dpi=900)
```

```python
usa = " ".join(review for review in wr[wr["country"]=="US"].description)
fra = " ".join(review for review in wr[wr["country"]=="France"].description)
ita = " ".join(review for review in wr[wr["country"]=="Italy"].description)
spa = " ".join(review for review in wr[wr["country"]=="Spain"].description)
por = " ".join(review for review in wr[wr["country"]=="Portugal"].description)

world= usa+fra+ita+spa+por
```

```python
plt.figure(figsize=(13,11))

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=60, 
                          random_state=42,
                          colormap="Accent",
                         ).generate(usa)

plt.imshow(wordcloud)
plt.title("USA", fontsize=40)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(13,11))

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=60, 
                          random_state=42,
                          colormap="tab20",
                         ).generate(fra)

plt.imshow(wordcloud)
plt.title("France", fontsize=40)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(13,11))

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=60, 
                          random_state=42,
                          colormap="Dark2"
                         ).generate(world)

plt.imshow(wordcloud)
plt.title("World", fontsize=40)
plt.axis('off')
plt.show()
```
